## 1.4 Cleaning our Dataset

## Twitter Dataset

```python
# converting the first row to be the column names
# 
twitter_data_2017.head(0)
header_row = 0

twitter_data_2017.columns = twitter_data_2017.iloc[header_row]
twitter_data_2017 = twitter_data_2017.drop(header_row)
```

```python
# replacing spaces in column names
twitter_data_2017.columns = twitter_data_2017.columns.str.replace(' ', '_').str.lower()
twitter_data_2017.head()
```

```python
# checking the data information
twitter_data_2017.info()
```

```python
twitter_data = twitter_data_2017.dropna(axis=1)
twitter_data.info()
```

```python
# Dropping of irrelevant Data Columns
#
twitter_data_2017 = twitter_data_2017.drop(['label', 'google_maps', 'bio', 'screen_name',	'full_name', 'tweet_id'], axis=1)
twitter_data_2017.head()
```

```python
twitter_data_2017.info()
```

```python
# converting the date column to date type
import datetime

date = '12/12/2017 15:33:16'
user_since = '6/16/2017'
twitter_data_2017.date = datetime.datetime.strptime(date, '%m/%d/%Y %H:%M:%S')
twitter_data_2017.user_since = datetime.datetime.strptime(user_since, '%m/%d/%Y')
```

```python
# Converting object values to integers
#
twitter_data_2017.followers = twitter_data_2017['followers'].astype(str).astype(int)
twitter_data_2017.follows =twitter_data_2017['follows'].astype(str).astype(int)
twitter_data_2017.retweets =twitter_data_2017['retweets'].astype(str).astype(int)
twitter_data_2017.favorites =twitter_data_2017['favorites'].astype(str).astype(int)
```

```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
twitter_data_2017['verfied']= label_encoder.fit_transform(twitter_data_2017['verfied'])
twitter_data_2017['profile_image']= label_encoder.fit_transform(twitter_data_2017['profile_image'])
```

```python
# Replace missing values with a number
twitter_data_2017['location'].fillna('unknown', inplace=True)
```

```python
#
#
twitter_data_2017[twitter_data_2017.columns[1]] = twitter_data_2017[twitter_data_2017.columns[1]].replace("[@$#:,.'/]", "", regex=True)
```

```python
twitter_data_2017.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
```
